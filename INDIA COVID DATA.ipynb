### sum and max

```python
total_cured = c_19_IND_df_copy.groupby('State/UnionTerritory')['Cured'].max().sum()
total_deaths = c_19_IND_df_copy.groupby('State/UnionTerritory')['Deaths'].max().sum()
total_confirmed = c_19_IND_df_copy.groupby('State/UnionTerritory')['Confirmed'].max().sum()
```

```python
print(f'Total cure numbers are {total_cured}')
print(f'Total Deaths due to covid 19 is {total_deaths}')
print(f'Total confirm people for covid 19 are {total_confirmed}')
```

### death rate

```python
death_rate = (total_deaths/total_confirmed)*100
print(death_rate)
```

```python
print('Death Rate of india for Covid 19 is {:.2f} %'.format(death_rate))
```

### active cases

```python
active_cases = total_confirmed - (total_cured + total_deaths)
```

```python
print(f'Total active cases for covid 19 are {active_cases}')
```

### recovery rate

```python
recovery_rate = (total_cured/ total_confirmed) * 100
print('Recovery Rate of india for Covid 19 is {:.2f} %'.format(recovery_rate))
```

```python
c_19_IND_df_copy['active_cases'] = (c_19_IND_df_copy.Confirmed - (c_19_IND_df_copy.Cured + c_19_IND_df_copy.Deaths))
```
